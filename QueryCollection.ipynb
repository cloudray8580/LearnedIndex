```python
import csv
def write_csv_file(path, head, data):  
    try:  
        with open(path, 'w', newline='') as csv_file:  
            writer = csv.writer(csv_file, dialect='excel')  
  
            if head is not None:  
                writer.writerow(head)  
  
            for row in data:  
                writer.writerow(row)  
  
            print("Write a CSV file to path %s Successful." % path)  
    except Exception as e:  
        print("Write an CSV file to path: %s, Case: %s" % (path, e)) 
```

```python
import pandas as pd
df = pd.read_csv("./Data/HilbertSortedPOIs.csv")
print(df)

LATITUDE_KM = 110.574*2
LONGITUDE_KM = 111.320*2

NUMBER_OF_REGIONS = 10

range_100m = []
range_200m = []
range_500m = []
range_1km = []
range_2km = []
range_5km = []
range_10km = []
range_20km = []
range_50km = []
range_100km = []

random_list = []
valid_count = 0

random.seed()
while valid_count < NUMBER_OF_REGIONS:
    position = random.randrange(len(df))
    x = df.iloc[position,0]
    y = df.iloc[position,1]
    if abs(x) >= 60:
        continue
    else:
        random_list.append([x,y])
        valid_count += 1

for item in random_list:
    x = item[0]
    y = item[1]
    range_100m.append([x-0.1/LATITUDE_KM,x+0.1/LATITUDE_KM,y-0.1/LONGITUDE_KM,y+0.1/LONGITUDE_KM])
    range_200m.append([x-0.2/LATITUDE_KM,x+0.2/LATITUDE_KM,y-0.2/LONGITUDE_KM,y+0.2/LONGITUDE_KM])
    range_500m.append([x-0.5/LATITUDE_KM,x+0.5/LATITUDE_KM,y-0.5/LONGITUDE_KM,y+0.5/LONGITUDE_KM])
    range_1km.append([x-1/LATITUDE_KM,x+1/LATITUDE_KM,y-1/LONGITUDE_KM,y+1/LONGITUDE_KM])
    range_2km.append([x-2/LATITUDE_KM,x+2/LATITUDE_KM,y-2/LONGITUDE_KM,y+2/LONGITUDE_KM])
    range_5km.append([x-5/LATITUDE_KM,x+5/LATITUDE_KM,y-5/LONGITUDE_KM,y+5/LONGITUDE_KM])
    range_10km.append([x-10/LATITUDE_KM,x+10/LATITUDE_KM,y-10/LONGITUDE_KM,y+10/LONGITUDE_KM])
    range_20km.append([x-20/LATITUDE_KM,x+20/LATITUDE_KM,y-20/LONGITUDE_KM,y+20/LONGITUDE_KM])
    range_50km.append([x-50/LATITUDE_KM,x+50/LATITUDE_KM,y-50/LONGITUDE_KM,y+50/LONGITUDE_KM])
    range_100km.append([x-100/LATITUDE_KM,x+100/LATITUDE_KM,y-100/LONGITUDE_KM,y+100/LONGITUDE_KM])
    
write_csv_file("./Data/RangeQueryCollection100m.csv",['lower_x','upper_x','lower_y','upper_y'],range_100m)
write_csv_file("./Data/RangeQueryCollection200m.csv",['lower_x','upper_x','lower_y','upper_y'],range_200m)
write_csv_file("./Data/RangeQueryCollection500m.csv",['lower_x','upper_x','lower_y','upper_y'],range_500m)
write_csv_file("./Data/RangeQueryCollection1km.csv",['lower_x','upper_x','lower_y','upper_y'],range_1km)
write_csv_file("./Data/RangeQueryCollection2km.csv",['lower_x','upper_x','lower_y','upper_y'],range_2km)
write_csv_file("./Data/RangeQueryCollection5km.csv",['lower_x','upper_x','lower_y','upper_y'],range_5km)
write_csv_file("./Data/RangeQueryCollection10km.csv",['lower_x','upper_x','lower_y','upper_y'],range_10km)
write_csv_file("./Data/RangeQueryCollection20km.csv",['lower_x','upper_x','lower_y','upper_y'],range_20km)
write_csv_file("./Data/RangeQueryCollection50km.csv",['lower_x','upper_x','lower_y','upper_y'],range_50km)
write_csv_file("./Data/RangeQueryCollection100km.csv",['lower_x','upper_x','lower_y','upper_y'],range_100km)
```
